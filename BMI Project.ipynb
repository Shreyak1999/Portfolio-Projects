```python
name = input("Enter your Name Here")

weight = int(input("Weight in kg: "))

feet = int(input("Height in feet: "))

inches = int(input("Height in inches: "))



BMI = (weight/(feet*0.3048 + inches*0.0254)**2)

BMI_your = round(BMI,2)

print(f"Your BMI is: {BMI_your}")

if BMI > 0:
    if BMI <= 18.5:
            print(name + ", you are Underweight with minimal health risks")
    elif BMI<=24.9:
            print(name + ", you have Normal Weight with minimal health risks")
    elif BMI<=29.9:
            print(name + ", you are Overweight with increased health risks")
    elif BMI<=39.9:
            print(name + ", you are Obese with high health risks")
    else: 
            print(name + ", you are Morbidly Obese with extremely high health risks")
```

Underweight = Below 18.5
Normal Weight = Between 18.6 to 24.9
Overweight = Between 25 to 29.9
Obese = Between 30 to 40
Morbidly Obese = Above 40
