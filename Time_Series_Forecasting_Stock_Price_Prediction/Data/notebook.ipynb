##### Libraries

```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = yf.download("AAPL", start="2015-01-01", end="2025-07-31", interval="1d")
print(df.head())
df.columns
```

```python
df = df['Close']['AAPL'].to_frame(name='price')
```

```python
df.tail()
```

```python
plt.figure(figsize=(12,6))
plt.plot(df.index, df['price'], label="AAPL Price")
plt.title("Apple Stock Price (2015-2025)")
plt.xlabel("Date")
plt.ylabel("Price (USD)")
plt.legend()
plt.show()
```
