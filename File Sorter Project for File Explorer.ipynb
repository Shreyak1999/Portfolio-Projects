```python
import os, shutil
```

```python
path = r"C:/Users/cysd1/OneDrive/Desktop/Stuff/"
```

```python
file_name = os.listdir(path)
```

```python
folder_name = ["pdf","excel","image","docs"]

for loop in range(0,4):
    folder_path = path + folder_name[loop]
    if not os.path.exists(folder_path):
       os.makedirs(folder_path)
```

```python
for file in file_name:
    if ".pdf" in file and not os.path.exists(path + "pdf/" + file):
        shutil.move(path + file, path + "pdf/" + file)
    elif ".jpg" in file and not os.path.exists(path + "image/" + file):
        shutil.move(path + file, path + "image/" + file)
    elif ".xlsx" in file and not os.path.exists(path + "excel/" + file):
        shutil.move(path + file, path + "excel/" + file)    
    elif ".docx" in file and not os.path.exists(path + "docs/" + file):
        shutil.move(path + file, path + "docs/" + file)  
    else:
        print("Nothing Left")
```
